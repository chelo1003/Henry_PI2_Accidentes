En el segundo dataset existen las siguientes columnas: 'moto', 'auto', 
'transporte_publico', 'camion', 'ciclista'. Voy a crear columnas de nulos en el primer dataset.

```python
df_homicidios["moto"] = np.empty(len(df_homicidios))
df_homicidios["auto"] = np.empty(len(df_homicidios))
df_homicidios["transporte_publico"] = np.empty(len(df_homicidios))
df_homicidios["camion"] = np.empty(len(df_homicidios))
df_homicidios["ciclista"] = np.empty(len(df_homicidios))
df_lesiones["rol"] = np.empty(len(df_lesiones))
df_lesiones = df_lesiones.drop("GRAVEDAD",axis=1)
```

```python
# Renombrar las columnas de los dataframes
df_homicidios = df_homicidios.rename(columns={"ID": "id"})
df_homicidios = df_homicidios.rename(columns={"N_VICTIMAS": "n_victimas"})
df_homicidios = df_homicidios.rename(columns={"FECHA": "fecha"})
df_homicidios = df_homicidios.rename(columns={"AAAA": "aaaa"})
df_homicidios = df_homicidios.rename(columns={"MM": "mm"})
df_homicidios = df_homicidios.rename(columns={"DD": "dd"})
df_homicidios = df_homicidios.rename(columns={"HORA": "hora"})
df_homicidios = df_homicidios.rename(columns={"HH": "franja_hora"})
df_homicidios = df_homicidios.rename(columns={"Dirección Normalizada": "direccion_normalizada"})
df_homicidios = df_homicidios.rename(columns={"LUGAR_DEL_HECHO": "direccion"})
df_lesiones = df_lesiones.rename(columns={"otra_direccion": "direccion"})
df_homicidios = df_homicidios.rename(columns={"TIPO_DE_CALLE": "tipo_calle"})
df_homicidios = df_homicidios.rename(columns={"Calle": "calle"})
df_homicidios = df_homicidios.rename(columns={"Altura": "altura"})
df_homicidios = df_homicidios.rename(columns={"Cruce": "cruce"})
df_homicidios = df_homicidios.rename(columns={"XY (CABA)": "geocodificacion_CABA"})
df_homicidios = df_homicidios.rename(columns={"COMUNA": "comuna"})
df_homicidios = df_homicidios.rename(columns={"pos x": "longitud"})
df_homicidios = df_homicidios.rename(columns={"pos y": "latitud"})
df_lesiones = df_lesiones.rename(columns={"latutid": "latitud"})
df_homicidios = df_homicidios.rename(columns={"PARTICIPANTES": "participantes"})
df_homicidios = df_homicidios.rename(columns={"VICTIMA_x": "victima"})
df_homicidios = df_homicidios.rename(columns={"ACUSADO": "acusado"})
df_homicidios = df_homicidios.rename(columns={"ROL": "rol"})
df_homicidios = df_homicidios.rename(columns={"VICTIMA_y": "vehiculo_victima"})
df_lesiones = df_lesiones.rename(columns={"VEHICULO_VICTIMA": "vehiculo_victima"})
df_homicidios = df_homicidios.rename(columns={"SEXO": "sexo"})
df_lesiones = df_lesiones.rename(columns={"SEXO": "sexo"})
df_homicidios = df_homicidios.rename(columns={"EDAD": "edad"})
df_lesiones = df_lesiones.rename(columns={"EDAD_VICTIMA": "edad"})
df_homicidios = df_homicidios.rename(columns={"FECHA_FALLECIMIENTO": "gravedad"})
```
